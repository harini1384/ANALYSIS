### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = df['mpg'].median()

above_median_mpg_cars = df[df['mpg'] > median_mpg].copy()  # Make an explicit copy

above_median_mpg_cars.loc[:, 'hp_to_weight'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

max_ratio = above_median_mpg_cars['hp_to_weight'].max()

best_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight'] == max_ratio]

print(best_cars.index.tolist())

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
df= df.reset_index(drop=False)

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Scatterplot
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(20, 200),
    alpha=0.7,
    palette='Set2'
)

# Titles and labels
plt.title('Horsepower vs Weight of Cars by Origin and MPG', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
